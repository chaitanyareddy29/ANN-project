```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data
data = pd.read_csv('Data.csv')

# create scatter plots between input variables and output variable
plt.scatter(data['AT'], data['PE'])
plt.xlabel('Temperature (AT)')
plt.ylabel('Electricity Output (PE)')
plt.title('Scatter Plot of Temperature vs. Electricity Output')
plt.show()

plt.scatter(data['V'], data['PE'])
plt.xlabel('Exhaust Vacuum (V)')
plt.ylabel('Electricity Output (PE)')
plt.title('Scatter Plot of Exhaust Vacuum vs. Electricity Output')
plt.show()

plt.scatter(data['AP'], data['PE'])
plt.xlabel('Ambient Pressure (AP)')
plt.ylabel('Electricity Output (PE)')
plt.title('Scatter Plot of Ambient Pressure vs. Electricity Output')
plt.show()

plt.scatter(data['RH'], data['PE'])
plt.xlabel('Relative Humidity (RH)')
plt.ylabel('Electricity Output (PE)')
plt.title('Scatter Plot of Relative Humidity vs. Electricity Output')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data
data = pd.read_csv('Data.csv')

# create a line graph of temperature and ambient pressure over time
plt.plot(data['AT'], label='Temperature')
plt.plot(data['AP'], label='Ambient Pressure')
plt.plot(data['V'], label='Vacume')
plt.plot(data['RH'], label='HUm')
plt.plot(data['PE'], label='PE')


plt.xlabel('Time')
plt.ylabel('Temperature and Pressure')
plt.title('Temperature and Pressure over Time')
plt.legend()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data.csv')

# Create a box plot
fig, ax = plt.subplots(figsize=(10, 8))
df.plot(kind='box', ax=ax)

# Set the plot title and axis labels
ax.set_title('Box plot of CCPP dataset')
ax.set_xlabel('Variable')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CCPP data into a DataFrame
ccpp = pd.read_csv('Data.csv')

# Compute the correlation matrix
corr = ccpp.corr()

# Create a heatmap using seaborn
sns.heatmap(corr, cmap='coolwarm', annot=True)

# Set plot title
plt.title('Heatmap')

# Show the plot
plt.show()
```
